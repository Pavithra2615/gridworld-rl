```python
import numpy as np

# Grid size
grid_size = 3
num_states = grid_size * grid_size
actions = [0, 1, 2, 3]  # up, down, left, right

gamma = 0.9

# Initialize policy and value function
policy = np.random.choice(actions, size=num_states)
V = np.zeros(num_states)

# Define goal and penalty states
goal_state = 8
penalty_state = 4

# Convert state to (row, col)
def get_position(state):
    return (state // grid_size, state % grid_size)

# Convert (row, col) to state
def get_state(pos):
    return pos[0] * grid_size + pos[1]

# Get next state
def get_next_state(state, action):
    row, col = get_position(state)

    if action == 0: row -= 1
    elif action == 1: row += 1
    elif action == 2: col -= 1
    elif action == 3: col += 1

    # Boundary check
    if row < 0 or row >= grid_size or col < 0 or col >= grid_size:
        return state

    return get_state((row, col))

# Reward function
def get_reward(state):
    if state == goal_state:
        return 10
    elif state == penalty_state:
        return -10
    else:
        return -1

# POLICY ITERATION
is_stable = False

while not is_stable:

    # Policy Evaluation
    while True:
        delta = 0
        for s in range(num_states):
            v = V[s]
            a = policy[s]
            next_s = get_next_state(s, a)
            r = get_reward(next_s)

            V[s] = r + gamma * V[next_s]
            delta = max(delta, abs(v - V[s]))

        if delta < 0.001:
            break

    # Policy Improvement
    is_stable = True

    for s in range(num_states):
        old_action = policy[s]
        action_values = []

        for a in actions:
            next_s = get_next_state(s, a)
            r = get_reward(next_s)
            action_values.append(r + gamma * V[next_s])

        best_action = np.argmax(action_values)
        policy[s] = best_action

        if old_action != best_action:
            is_stable = False

# Output
print("Optimal Policy:")
for s in range(num_states):
    print(f"State {s} -> Action {policy[s]}")

print("\nValue Function:")
print(V)
```
